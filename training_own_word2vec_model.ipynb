```python
import zipfile
zip_ref = zipfile.ZipFile('/content/imdb-dataset-of-50k-movie-reviews.zip', 'r')
zip_ref.extractall('/content')
zip_ref.close()
```

```python
import pandas as pd
df=pd.read_csv(r'/content/imdb-dataset-of-50k-movie-reviews.zip')
df
```

```python
import gensim
```

```python
df['review']=df['review'].apply(gensim.utils.simple_preprocess)
```

```python
df
```

```python
import gensim
```

```python
model=gensim.models.Word2Vec(
    window=10,
    min_count=2,
)
```

```python
model.build_vocab(df['review'])
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.train(df['review'],total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar('man')
```

```python
model.wv.most_similar('movie')
```

```python
model.wv.doesnt_match(['acting','movie','donkey','school'])
```

```python
model.wv['man']
```
