```python
import pandas as pd
df=pd.read_csv(r'/content/imdb-dataset-of-50k-movie-reviews.zip')
df
```

```python
import gensim
```

```python
df['review']=df['review'].apply(gensim.utils.simple_preprocess)
```

```python
import gensim
```

```python
model=gensim.models.Word2Vec(
    window=10,
    min_count=2,
)
```

```python
model.build_vocab(df['review'])
```

```python
model.train(df['review'],total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar('man')
```

```python
# lets do averge word2vec for sentence
# till now we have words vector , now we want  a vector of review
```

```python
import numpy as np
import pandas  as pd
```

```python
def document_vector(doc):
  doc=[word for word in doc if word  in model.wv.index_to_key]
  return np.mean(model.wv[doc],axis=0)
```

```python
document_vector(df['review'].values[0])
```

```python
from tqdm import tqdm
```

```python
X=[]
for doc in tqdm(df['review'].values):
  X.append(document_vector(doc))
```

```python
# in above code we are passing every review and converting them to vector
```

```python
X=np.array(X)
```

```python
X.shape
```

converting true/false to 0/1

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

y=encoder.fit_transform(df['sentiment'])
```

```python
y
```

```python
from  sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=56)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
accuracy_score(y_test,y_pred)
```
